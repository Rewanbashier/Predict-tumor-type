```python
#read dataset
import pandas as pd 
pd.options.display.max_rows=None
pd.options.display.float_format = '{:,.2f}'.format
```

```python
data=pd.read_csv(r'D:/Data Analysis/The-Pandas-Workshop-master/KNNAlgorithmDataset.csv')
```

```python
#drop unwanted column
data1=data.drop(['id'],axis=1)
```

```python
data1
```

```python
data1.info()
```

```python
#input values
x=data1.iloc[:,1:]
```

```python
#output value (Benign or malignant tumor)
y=data1.iloc[:,0]
```

```python
# label encoding to convert output(Benign or malignant tumor)
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
y=encoder.fit_transform(y)
```

```python
#split dataset into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=44)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kr=KNeighborsClassifier(n_neighbors=3)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
imputing=SimpleImputer(missing_values=np.nan,strategy='mean')
x_train=imputing.fit_transform(x_train)
x_test=imputing.fit_transform(x_test)
```

```python
kr.fit(x_train,y_train)
```

```python
y_pred=kr.predict(x_test)
```

```python
kr.score(x_train,y_train)*100
```

```python
kr.score(x_test,y_test)*100
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
# apply Bernoulli naive bayes
from sklearn.naive_bayes import BernoulliNB
```

```python
model=BernoulliNB(alpha=1.0,binarize=1)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred2=model.predict(x_test)
```

```python
print(model.score(x_train,y_train)*100)
```

```python
print(model.score(x_test,y_test)*100)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred2)*100)
```
